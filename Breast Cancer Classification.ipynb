# Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Collection and processing

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
#number of rows and columns
df.shape
```

```python
#some more info regarding the data
df.info()
```

```python
#checking missing values
df.isnull().sum()
```

```python
#statistical measures about the data
df.describe()
```

```python
#checking the distribution of the diagnosis value
df['diagnosis'].value_counts()
```

```python
df.replace({"diagnosis":{"M":0,"B":1}},inplace=True)
```

```python
df.head()
```

# Seperating the features and the target

```python
x=df.drop(columns=['id','diagnosis'],axis=1)
x
```

```python
y=df['diagnosis']
y
```

# Splitting the training and testing datset

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

# Model Training

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train , y_train)
```

# Model Evaluation

```python
#accuracy on training data
x_training_data=lr.predict(x_train)
training_accuracy_score=accuracy_score(y_train,x_training_data)
print("Accuracy on training data:",training_accuracy_score)
```

```python
#accuracy on testing data
x_testing_data=lr.predict(x_test)
testing_accuracy_score=accuracy_score(y_test,x_testing_data)
print("Accuracy on training data:",testing_accuracy_score)
```
