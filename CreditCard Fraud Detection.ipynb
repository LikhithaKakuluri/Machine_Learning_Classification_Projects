# Importing the Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

# Data Collection and preprocessing

```python
df=pd.read_csv("creditcard.csv")
df
```

```python
#Statistical mreasures
df.describe()
```

```python
#no. of rows and columns
df.shape
```

```python
#checking missing value
df.isnull().sum()
```

```python
#some more info regarding datset
df.info()
```

```python
df['Class'].value_counts()
```

This Dataset is highly unbalanced

0--->Normal Transaction

1--->Fraudulent Transaction

```python
#Seperating the data for analysis
normal=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(normal.shape)
print(fraud.shape)
```

```python
#Statistical measures of data
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the values of both the transaction
df.groupby('Class').mean()
```

# Under Sampling

Build a sample dataset containing similar distribution of normal & Fraudulent Transaction

```python
normal_sample=normal.sample(n=492)
```

# Concatenating two DataFrames

```python
df1=pd.concat([normal_sample,fraud],axis=0)
```

```python
df1
```

```python
df1['Class'].value_counts()
```

# Dividing the feature and labels

```python
x=df1.drop('Class',axis=1)
x
```

```python
y=df1['Class']
y
```

# Splitting the training and testing dataset

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

# Model Evaluation

```python
x_training_data=lr.predict(x_train)
training_data_accuracy=metrics.accuracy_score(y_train,x_training_data)
print("Accuracy of training dataset:",training_data_accuracy)
```

```python
x_testing_data=lr.predict(x_test)
testing_data_accuracy=metrics.accuracy_score(y_test,x_testing_data)
print("Accuracy of testing dataset:",testing_data_accuracy)
```
